# FUZZY CLUSTERING APPROACH (1543 songs, 10 Clusters)

```python
# Specify INPUT songs list

input_songs_data = df.loc[[360, 600, 740, 960, 1200, 1370, 1371, 1372, 1373, 1374, 1375, 1490]]
print('PLAYLIST OF INPUT SONGS ARE:')
input_songs_data['playlist_name']
```

```python
# Fuzzy Clustering algorithm

import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import skfuzzy as fuzz

# Step 1: Load the song data
song_data = df

# Step 2: Define the input songs
input_songs = ['White Christmas', 'Let It Snow! Let It Snow! Let It Snow!', 'Jingle Bell Rock', 'Winter Wonderland']

# Step 4: Preprocess numerical features
numerical_features = ['artist_fraction', 'playlist_description_encoded']  # Replace with the actual numerical feature column names
all_songs_numerical = song_data[numerical_features]
input_songs_numerical = input_songs_data[numerical_features]

# Step 5: Preprocess categorical features
categorical_features = ['track_track_name', 'playlist_name']#, 'playlist_description']  # Replace with the actual categorical feature column names
all_songs_categorical = song_data[categorical_features]
input_songs_categorical = input_songs_data[categorical_features]

# Step 6: Perform one-hot encoding on categorical features
encoder = OneHotEncoder(sparse_output=False)
all_songs_categorical_encoded = encoder.fit_transform(all_songs_categorical)
input_songs_categorical_encoded = encoder.transform(input_songs_categorical)


# Step 7: Combine numerical and categorical features
all_songs_combined = np.concatenate((all_songs_numerical, all_songs_categorical_encoded), axis=1)
input_songs_combined = np.concatenate((input_songs_numerical, input_songs_categorical_encoded), axis=1)

# Step 8: Scale the feature values
scaler = MinMaxScaler()
X_combined = scaler.fit_transform(all_songs_combined)
input_songs_combined_scaled = scaler.transform(input_songs_combined)

# Step 9: Train the fuzzy clustering model
num_clusters = 10
cntr, _, _, _, _, _, _ = fuzz.cluster.cmeans(X_combined.T, num_clusters, m=1.001, error=0.00005, maxiter=50000)

# Step 10: Predict the clusters for all songs
all_songs_cluster_membership = np.argmax(fuzz.cluster.cmeans_predict(X_combined.T, cntr, 2, error=0.00005, maxiter=50000)[0], axis=0)

# Step 11: Predict the clusters for the input songs
input_cluster_membership = np.argmax(fuzz.cluster.cmeans_predict(input_songs_combined_scaled.T, cntr, 2, error=0.00005, maxiter=50000)[0], axis=0)

# Step 12: Get recommendations from the selected clusters
recommendations = []
for cluster in np.unique(input_cluster_membership):
    cluster_songs = song_data[all_songs_cluster_membership == cluster]
    recommendations.extend(cluster_songs['track_track_name'].tolist())

# Step 13: Select the best matching 6th song recommendation
song_6_recommendation = max(set(recommendations), key=recommendations.count)

#print("Recommended next song:", song_6_recommendation)
```

# Best Song using Euclidean Distance

```python
from scipy.spatial.distance import cdist
```

```python
input_maxcount_cluster = int(np.median(input_cluster_membership))
#input_maxcount_cluster
```

```python
# Calculate Eucledian distance (all songs to selected cluster)
distances = cdist(X_combined, cntr, metric='euclidean')
```

```python
# index of the song with minimum distance to the center of the selected cluster
allsong_distance_to_selectedcluster = (distances[:, input_maxcount_cluster])
min_index = np.argmin(allsong_distance_to_selectedcluster)
```

```python
# song with the calculated index
next_song = df['track_track_name'][min_index]
print('RECOMENDED SONG:', next_song)
```

```python
# playlist of the predicted song
filtered_df = df[df['track_track_name'] == next_song]
print('PLAYLIST OF THE RECOMENDED SONG:', filtered_df['playlist_name'])
```
